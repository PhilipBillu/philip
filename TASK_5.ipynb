**ROAD CONDITION**

```python
#Finding Mean of Road_Condition_Class
df['Road_Condition_Class'].mean()
```

```python
#Finding Median for Road_Condition_Class
df['Road_Condition_Class'].median()
```

```python
#Finding Mode for Road_Condition_Class
df['Road_Condition_Class'].mode()
```

```python
#Finding Standard Deviation for Road_Condition_Class
```

```python
Sub_Deviation = df['Road_Condition_Class']-df['Road_Condition_Class'].mean()
```

```python
Square_Deviation = (df['Road_Condition_Class']-df['Road_Condition_Class'].mean())**2
```

```python
Add_Deviation=0
for i in Square_Deviation:
  Add_Deviation=Add_Deviation+i
print("ADD_Deviation : ",Add_Deviation)
```

```python
Div_Deviation = Add_Deviation / len(df['Road_Condition_Class'])
```

```python
Standard_Deviation = Div_Deviation**0.5
print("Standard Deviation ",Standard_Deviation)
```
